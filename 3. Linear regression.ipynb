```python
#LR model written in LinearRegressionModel.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model  
```

```python
df = pd.read_csv('weight.csv')
df.head(11)
```

```python
df.shape
```

```python
df.plot(kind='scatter', x='Waist_cm', y='Weight_kg')
plt.show()
```

```python
df.plot(kind='box')
plt.show()
```

```python
df.corr()#Correlation Coefficients
```

```python
#Change to dataframe variables
Waist=pd.DataFrame(df['Waist_cm'])
Weight=pd.DataFrame(df['Weight_kg'])
```

```python
Waist
```

## Build Linear regression model

```python
lm=linear_model.LinearRegression()
model=lm.fit(Waist, Weight)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(Waist, Weight)#Evaluate the model
```

## Predict new values of Weight

```python
X=([67,78,98])
X=pd.DataFrame(X)
Y=model.predict(X)
Y=pd.DataFrame(Y)
df1=pd.concat([X, Y], axis=1, keys=['Waist_new', 'Weight_Predicted'])
df1
```

```python
#Visualize the result
df.plot(kind='scatter', x='Waist_cm', y='Weight_kg')

#Plotting the Regression Line
plt.plot(Waist,model.predict(Waist), color='red', linewidth=2)
plt.show()
```
